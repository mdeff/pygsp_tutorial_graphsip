```python
import numpy as np
import scipy
import pygsp
from matplotlib import pyplot as plt
import pandas as pd
```

```python
W = scipy.sparse.rand(10, 10, 0.2)
W = W.toarray()
W = W - np.diag(np.diag(W))
W = W + np.transpose(W)
G = pygsp.graphs.Graph(W)
degree_vector = G.d
print(degree_vector)
```

```python
G = pygsp.graphs.Logo()
G.estimate_lmax()
g = pygsp.filters.Heat(G, tau=100)
DELTAS = [20, 30, 1090]
s = np.zeros(G.N)
s[DELTAS] = 1
s = g.filter(s)
G.plot_signal(s, highlight=DELTAS, backend='matplotlib')
```
